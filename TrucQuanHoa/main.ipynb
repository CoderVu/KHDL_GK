```python

import pandas as pd
import matplotlib.pyplot as plt

```

```python
# Đọc dữ liệu từ file CSV
df = pd.read_csv("D:\DUT\DH_HK6\KHDL\DEALINEGK\TienXuLyData\Processed_data.csv")
```

```python

# Tính toán số lượng cầu thủ từng quốc gia
country_counts = df['Country'].value_counts()

# Chọn số lượng quốc gia cần trực quan hóa (ví dụ: 10 quốc gia)
top_countries = country_counts.head(10)

# Tạo biểu đồ cột
plt.figure(figsize=(10, 6))
top_countries.plot(kind='bar', color='skyblue')
plt.xlabel('Country')
plt.ylabel('Number of Players')
plt.title('Top 10 Countries with Most ATP Players')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
